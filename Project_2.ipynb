```python
import re
```

```python
df['Rotten Tomatoes_%']=0.0
for i in range(len(df)):
    df['Rotten Tomatoes_%'][i]=float(re.search(r'\d*', df['Rotten Tomatoes'][i])[0])
df
```

```python
df['IMDb_%']=0.0
df['CombRating']=0.0
for i in range(len(df)):
    try:
        df['IMDb_%'][i]=float(re.search(r'\d.\d', df['IMDb'][i])[0])*10
        df['CombRating'][i]=(df['Rotten Tomatoes_%'][i]+df['IMDb_%'][i])/2
    except:
        df['IMDb_%'][i]= None
        df['CombRating'][i]= None
df
```

```python
df['min_age']=0
for i in range(len(df)):
    try:
        df['min_age'][i]=int(re.search(r'\d*', df['Age'][i])[0])
    except:
        df['min_age'][i]= None
df
```
